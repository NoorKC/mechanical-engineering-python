### Newton Raphson Numerical Method

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def f(x):
    return x**3 - x - 1

def g(x):
    return derivative(f,x,dx=1e-4)

x0 = 1
error = 1e-5

while f(x)>error:
    x1 = x0 - (f(x0)/g(x0))
    x0 = x1
    
    print(x1)

x = np.linspace(-3,3,30)
plt.plot(x,f(x))
plt.grid(linestyle='--',color='black')
```

### Optimization - Minima of Function

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

def f(x):
    return np.cos(x) - 3*np.exp(-(x-0.2)**2)

in_x1 = 1 # starting initial guess
min_x1 = fmin(f,in_x1)
in_x2 = 2
min_x2 = fmin(f,in_x2)

plt.figure(figsize=(10,5))
x = np.linspace(-10,10,1000)
plt.plot(x,f(x),color='black',linestyle='--',label='$cos(x) - 3e^-(x-0.2)^2$') # function curve
plt.plot(in_x1,f(in_x1),'or',label='$X1 = {}$'.format(in_x1))
plt.plot(min_x1,f(min_x1),'vr',label='$min_x1={}$'.format(min_x1))

plt.plot(in_x2,f(in_x2),'og',label='$X2 = {}$'.format(in_x2))
plt.plot(min_x2,f(min_x2),'vg',label='$min_x2={}$'.format(min_x2))

plt.legend()
plt.grid(color='grey',linestyle='--',alpha=0.5)
plt.xlim(x[0],x[-1])
plt.show()
```

### Linear Programming Problem - Simplex Method

```python
import numpy as np
from scipy.optimize import linprog

Z = np.array([5,7,12]) # objective function

# inequality constraints

A = np.array([[-1,-1,-1],
              [-1,2,0],
              [0,0,-1]])

B = np.array([-500,0,-340])

x_low = 5
x_upper = 500

result = linprog(Z,A,B, bounds=[x_low,x_upper])

print('Minimum value of Objective Function',result.fun)
print('Value of x1,x2,x3: ',result.x)
```
